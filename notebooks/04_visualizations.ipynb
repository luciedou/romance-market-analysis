```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les donnes
df = pd.read_csv("../data/processed/romance_with_subgenres.csv", 
                 encoding='utf-8', sep=';')

# Configuration style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# ========================================
# GRAPHIQUE 1 : Corrlation avis/notes
# ========================================
plt.figure(figsize=(10, 6))
plt.scatter(df['ratings_count'], df['average_rating'], alpha=0.5)
plt.xlabel('Nombre d\'avis', fontsize=12)
plt.ylabel('Note moyenne', fontsize=12)
plt.title('Corrlation entre nombre d\'avis et note moyenne', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../visuals/correlation_avis_notes.png', dpi=300, bbox_inches='tight')
plt.show()

# Calculer la corrlation
corr = df['ratings_count'].corr(df['average_rating'])
print(f"Corrlation : {corr:.3f}")

# ========================================
# GRAPHIQUE 2 : Notes moyennes par sous-genre
# ========================================
subgenres_data = {
    'Historical': df[df['is_historical']]['average_rating'].mean(),
    'Contemporary': df[df['is_contemporary']]['average_rating'].mean(),
    'Paranormal': df[df['is_paranormal']]['average_rating'].mean(),
    'Fantasy': df[df['is_fantasy']]['average_rating'].mean(),
    'Erotic': df[df['is_erotic']]['average_rating'].mean()
}

plt.figure(figsize=(10, 6))
bars = plt.bar(subgenres_data.keys(), subgenres_data.values(), color=['#8B4513', '#4682B4', '#9370DB', '#FF69B4', '#DC143C'])
plt.ylabel('Note moyenne', fontsize=12)
plt.title('Notes moyennes par sous-genre de romance', fontsize=14, fontweight='bold')
plt.ylim(3.5, 4.1)

# Ajouter les valeurs sur les barres
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('../visuals/notes_par_subgenre.png', dpi=300, bbox_inches='tight')
plt.show()

# ========================================
# GRAPHIQUE 3 : Engagement par sous-genre
# ========================================
engagement_data = {
    'Historical': df[df['is_historical']]['ratings_count'].mean(),
    'Contemporary': df[df['is_contemporary']]['ratings_count'].mean(),
    'Paranormal': df[df['is_paranormal']]['ratings_count'].mean(),
    'Fantasy': df[df['is_fantasy']]['ratings_count'].mean(),
    'Erotic': df[df['is_erotic']]['ratings_count'].mean()
}

plt.figure(figsize=(10, 6))
bars = plt.bar(engagement_data.keys(), engagement_data.values(), color=['#8B4513', '#4682B4', '#9370DB', '#FF69B4', '#DC143C'])
plt.ylabel('Nombre moyen d\'avis', fontsize=12)
plt.title('Engagement par sous-genre (nombre d\'avis)', fontsize=14, fontweight='bold')

# Ajouter les valeurs sur les barres
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:,.0f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('../visuals/engagement_par_subgenre.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n Tous les graphiques sont sauvegards dans visuals/ !")
```
