```python
import pandas as pd

# Charger les données nettoyées
df_romance = pd.read_csv("../data/processed/romance_books_clean.csv", 
                         encoding='utf-8',
                         sep=';')

print(f"Dataset chargé : {len(df_romance)} romances\n")

# IDENTIFIER LES SOUS-GENRES

# Créer des colonnes booléennes pour chaque sous-genre
df_romance['is_historical'] = df_romance['genres'].str.contains('Historical', case=False, na=False)
df_romance['is_contemporary'] = df_romance['genres'].str.contains('Contemporary', case=False, na=False)
df_romance['is_paranormal'] = df_romance['genres'].str.contains('Paranormal', case=False, na=False)
df_romance['is_fantasy'] = df_romance['genres'].str.contains('Fantasy', case=False, na=False)
df_romance['is_erotic'] = df_romance['genres'].str.contains('Erotic', case=False, na=False)

# Compter combien de livres par sous-genre
print("=== RÉPARTITION PAR SOUS-GENRE ===")
print(f"Historical Romance : {df_romance['is_historical'].sum()} livres ({df_romance['is_historical'].sum()/len(df_romance)*100:.1f}%)")
print(f"Contemporary Romance : {df_romance['is_contemporary'].sum()} livres ({df_romance['is_contemporary'].sum()/len(df_romance)*100:.1f}%)")
print(f"Paranormal Romance : {df_romance['is_paranormal'].sum()} livres ({df_romance['is_paranormal'].sum()/len(df_romance)*100:.1f}%)")
print(f"Fantasy Romance : {df_romance['is_fantasy'].sum()} livres ({df_romance['is_fantasy'].sum()/len(df_romance)*100:.1f}%)")
print(f"Erotic Romance : {df_romance['is_erotic'].sum()} livres ({df_romance['is_erotic'].sum()/len(df_romance)*100:.1f}%)")

# Statistiques par sous-genre
print("\n=== NOTES MOYENNES PAR SOUS-GENRE ===")
print(f"Historical : {df_romance[df_romance['is_historical']]['average_rating'].mean():.2f}")
print(f"Contemporary : {df_romance[df_romance['is_contemporary']]['average_rating'].mean():.2f}")
print(f"Paranormal : {df_romance[df_romance['is_paranormal']]['average_rating'].mean():.2f}")
print(f"Fantasy : {df_romance[df_romance['is_fantasy']]['average_rating'].mean():.2f}")
print(f"Erotic : {df_romance[df_romance['is_erotic']]['average_rating'].mean():.2f}")

print("\n=== ENGAGEMENT MOYEN (nombre d'avis) PAR SOUS-GENRE ===")
print(f"Historical : {df_romance[df_romance['is_historical']]['ratings_count'].mean():.0f} avis")
print(f"Contemporary : {df_romance[df_romance['is_contemporary']]['ratings_count'].mean():.0f} avis")
print(f"Paranormal : {df_romance[df_romance['is_paranormal']]['ratings_count'].mean():.0f} avis")
print(f"Fantasy : {df_romance[df_romance['is_fantasy']]['ratings_count'].mean():.0f} avis")
print(f"Erotic : {df_romance[df_romance['is_erotic']]['ratings_count'].mean():.0f} avis")

# Sauvegarder
df_romance.to_csv("../data/processed/romance_with_subgenres.csv", index=False, sep=';', encoding='utf-8')
print("\n✅ Données avec sous-genres sauvegardées !")
```
